```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cobra
from MatplotProp import CanvasStyle, PltProps
PltProps()
```

```python
# all_slope = pd.read_csv('FlyDFA_slope.csv')
t_points = [0, 1, 2, 3, 4, 5, 6, 8, 10, 12, 14, 16, 18]
files = ['Proteome_CFR/Fly_Proteome_{}.csv'.format(t) for t in t_points]

data = pd.read_csv(files[0])
res = np.zeros((len(data), len(t_points)))
names = []
for i, file in enumerate(files):
    data = pd.read_csv(file)
    for j in range(data.shape[0]):
        row = data.iloc[j]
        read = str(row.values[0]).split(";")
        if i==0:
            names.append(read[0])
        res[j, i] = read[1]
            
df = pd.DataFrame(data=res)
df.columns = t_points
df.index = names
```

```python
df
```

```python
fig = plt.figure(figsize=(13, 8))
ax = fig.add_subplot(111)
sns.heatmap(df)
```
